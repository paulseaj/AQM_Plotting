```python
#@title Local Maxima and Local Minima plotted peaks and troughs
data = d1['pm1'].values
doublediff = np.diff(np.sign(np.diff(data)))
peak_locations = np.where(doublediff == -2)[0] + 1

doublediff2 = np.diff(np.sign(np.diff(-1*data)))
trough_locations = np.where(doublediff2 == -2)[0] + 1
plt.figure(figsize=(16,10), dpi= 80)
plt.plot('timestamp' ,'pm1', data=d1, color='tab:blue', label='Air Traffic')
plt.scatter(d1.timestamp[peak_locations], d1.pm1[peak_locations], marker=mpl.markers.CARETUPBASE, color='tab:green', s=100, label='Peaks')
plt.scatter(d1.timestamp[trough_locations], d1.pm1[trough_locations], marker=mpl.markers.CARETDOWNBASE, color='tab:red', s=100, label='Troughs')
```

```python
for t, p in zip(trough_locations[1::5], peak_locations[::3]):
    plt.text(d1.timestamp[p], d1.pm1[p]+15, d1.timestamp[p], horizontalalignment='center', color='darkgreen')
    plt.text(d1.timestamp[t], d1.pm1[t]-35, d1.timestamp[t], horizontalalignment='center', color='darkred')
```
